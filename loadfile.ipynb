```python
import h5py
import numpy as np

f = h5py.File('/media/xian/Data/dataset/Bekaert/h5files/42.h5', 'r')
dset = f['time_data']
multi_signal = np.empty(dset.shape, dtype=float)
for ind in np.arange(dset.shape[1]):
    multi_signal[:,ind] = dset[:,ind]
fs = 51200
time = np.linspace(0, multi_signal.shape[0] / fs, num=multi_signal.shape[0])
```

```python
import matplotlib.pyplot as plt

plt.figure(1)
plt.title("Signal Wave...")
plt.plot(time, multi_signal[:,0])
plt.show()
```

```python
from scipy.io.wavfile import write

channel = 1
enhanced_audio = multi_signal[:,channel] / np.max(np.abs(multi_signal[:,channel])) * 0.7
write("42.wav", fs, enhanced_audio.astype(np.float32))
```
