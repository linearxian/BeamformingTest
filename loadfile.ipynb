```python
import h5py
import numpy as np

f = h5py.File('test_signal_9_left_middle_right.h5', 'r')
dset = f['time_data']
dset.shape
signal = dset[:,0]
fs = 51200
time = np.linspace(0, len(signal) / fs, num=len(signal))
```

```python
import matplotlib.pyplot as plt

plt.figure(1)
plt.title("Signal Wave...")
plt.plot(time, signal)
plt.show()
```
